## Corrections <a class="anchor" id="1.1"></a>

```python
# Standerdize column names
train.columns = format_column_name(train)
X_test.columns = format_column_name(X_test)
y_test.columns = format_column_name(y_test)

# Calculate age from date of birth
train = calculate_age_from_dob(train, 'date_of_birth')
X_test = calculate_age_from_dob(X_test, 'date_of_birth')
```

```python
# Correct misrepresented values
train['gender'].replace('P', np.nan, inplace=True)
X_test['gender'].replace('P', np.nan, inplace=True)

train['healthcare_access'].replace('?', np.nan, inplace=True)
X_test['healthcare_access'].replace('?', np.nan, inplace=True)

train.loc[train['healthcare_costs'] < 0, 'healthcare_costs'] = np.nan
X_test.loc[X_test['healthcare_costs'] < 0, 'healthcare_costs'] = np.nan

train['urban_or_rural'].replace('rural', 'Rural', inplace=True)
X_test['urban_or_rural'].replace('rural', 'Rural', inplace=True)

train['urban_or_rural'].replace('urban', 'Urban', inplace=True)
X_test['urban_or_rural'].replace('urban', 'Urban', inplace=True)
```

```python
# Drop constant columns
train.drop(columns=['marital_status', 'transfusion_history'], inplace=True)
X_test.drop(columns=['marital_status', 'transfusion_history'], inplace=True)
```

```python
# Round all numeric values to 0 decimal places
numeric, categorical = datatype_distinction(train)
train[numeric.columns] = train[numeric.columns].round(0)
X_test[numeric.columns] = X_test[numeric.columns].round(0)
```
